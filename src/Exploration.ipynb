If values for a stock are missing for some days, we can consider:
* Ignoring the values imply that the returns will be computed between larger periods of time than for other values.
* Forward filling the values (and thus assuming the value of the stock remains constant) will result in zero returns for some days.
* Using a linear interpolation. This propagates some information from the future backward, and skews the behavior of the return for those days.
* Use stock known to be similar to infer what may have happened and impute missing values.

# Compute betas

We now compute the return and the corresponding betas for a rolling window. We take 15 days window here though this is a small history and makes the betas more sensitive to outliers.

```python
%%time

df = Betas.attach_return(data, "symbol")
df_ref = Betas.attach_return(data_ref, "symbol")

df0 = df.reset_index().merge(
    df_ref.reset_index().drop("symbol", axis=1),
    on="date",
    suffixes=["", "_ref"]
).dropna().set_index(df.index.names)
```

```python
%%time
df = Betas.attach_beta(df0, "return_ref", "return", "symbol", 15, kind="numpy")
```

```python
%%time
df_ols = Betas.attach_beta(df0, "return_ref", "return", "symbol", 15, kind="ols")
```

```python
%%time
df_byhand = Betas.attach_beta(df0, "return_ref", "return", "symbol", 15, kind="byhand")
```

```python
df.sample(10)
```

For faster implementation, here are some alternative to consider:

* Numpy has an optimized workflow for matrix computation. We can leverage that by computing the beta as $Cov(r, r_{ref})/Var(r_{ref})$.
* Each regression can be computed in parallel. Dask is a library that implements numpy/pandas in parallel, and loads in memory the data as needed.
* Each security can be computed in parallel, and the least square solution for each security can be found by gradient descent using an exponentially decaying momentum in the gradient, by going through the points in a chronological order. This has the advantage that the data only has to be read once, and does not need to be loaded completely in memory, but this may affect the betas obtained.

Given the large variation in the beta, we can regularize the regression to reduce the impact of outliers. The amount of regularization is controled by the parameter alpha from the Ridge regression. The betas are sensitive to the value selected.

```python
%%time
df_ridge = Betas.attach_beta(df0, "return_ref", "return", "symbol", 15, kind="ridge", alpha=0.1)
```

# Beta Hedging

We want to have a portfolio which is beta neutral, i.e. a portfolio who's average beta weighted by the amount of each stock owned gives zero. Given that we want to hedge against the index, we need to build a portfolio of stocks that we expect will move in a direction opposite to the index (whose beta is 1) over a long time (21 days), so that we can rebalance the portfolio every day with the index.

To measure the performance of a beta hedging strategy, we can 
* look at how successful we were at maintaining beta-neutrality of the portfolio, by measuring the average beta belance of the portfolio over the 21 days,
* compute the profit generated at the end of the 21 days.

Here we look at a portfolio with one stock (and the index). We assume we have a budget large enough to buy the desired quantity of the index on any given day. We can thus always rebalance as long as the stock has a negative beta by buying enough of the index. We simply need to measure the average days when the beta was positive. The smaller the value is, the more successful we were at balancing.

```python
n = 21
df[["beta"]].assign(floor=0.).max(axis=1).groupby(level="symbol", group_keys=False).rolling(n).mean().dropna()
```
